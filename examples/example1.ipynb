Example 1 - Calculating Extinction
----------------------------------

This notebook gives a simple example showing how to use OpenModes. A pair of split-ring resonators are created, and their extinction cross section is calculated

First we load import the `openmodes` packages, along with a few other useful python packages.

```python
# allow plots within the notebook
%run plot_settings.ipy
%pylab inline

# import useful python libraries
import os.path as osp

# import the openmodes packages
import openmodes
from openmodes.visualise import plot_parts
from openmodes.constants import c, eta_0
```

Then we create a `Simulation` object which holds all the data for the simulation we are going to run

```python
sim = openmodes.Simulation(name='example1')
```

Next we load the geometry file. This geometry file is written in the scripting language of [gmsh](http://geuz.org/gmsh), a program which converts the geometry into a surface mesh. Here we will use one of the geometry files provided with openmodes. 

The mesh density may be specified in the geometry file, but can be over-ridden with the parameter `mesh_tol`. Be careful with this parameter, setting it too small can result in very long computations.

```python
# if gmsh is not installed, replace "SRR.geo" with "SRR.msh" to use the precalculated mesh file
filename = osp.join(openmodes.geometry_dir, "SRR.geo")
mesh_tol = 0.5e-3
srr = sim.load_mesh(filename, mesh_tol)
```

Now we place the parts within the simulation. By default the parts are placed at the origin, so after placement we need to move one of them to the desired distance from the other. Before moving the second ring, it is rotated $180^\circ$ to create a broadside-coupled configuration. 

```python
srr1 = sim.place_part(srr)
srr2 = sim.place_part(srr)
srr2.rotate(axis = [0, 0, 1], angle = 180)
srr2.translate([0, 0, 2e-3])
```

To check that the parts have been placed in the correct location, we can visualise them using the provided function. This creates a crude 3D plot, with a cross at each of the nodes of the mesh. OpenModes also allows you to produce higher quality 3D plots by exporting to the VTK file format. However, as a quick check, the function `plot_parts` is much simpler, and can be used within a notebook. You can change the perspective by modifying the `view_angles` parameter. Note that the different axes have different scales.

```python
plot_parts(sim.parts, view_angles=(50, 70))
```

We will calculate the extinction cross section for this pair of SRRs, assuming that they are perfectly conducting. We want to excite it with a $y$ polarised plane wave, propagating in the $x$ direction. We will calculate at 201 frequencies between 4.5 and 8 GHz.

```python
freqs = linspace(4.5e9, 8e9, 401)
e_inc = array([0, 1, 0])
k_hat = array([1, 0, 0])

# create an empty array to hold extinction data
extinction_single = empty(len(freqs), float64)
extinction_pair = empty(len(freqs), float64)
```

Now loop through all the frequencies. For notational convenience, the time dependence is assumed as $\exp(s t)$ with complex frequency $s$.

At each frequency the impedance matrix `Z` is calculated, as is the source term `V` due to the plane wave. By default these matrices and vectors are composite objects accounting for all the parts. We can select one of these matrices to find the response of an isolated part, or we can combine them together to find the response of the entire system.

In both cases, the extinction cross-section $\sigma_{ext}$ is found from the impedance matrix $Z$ and driving term $V$ as $\sigma_{ext}(s) = V^{*}(s)\cdot Z(s) \cdot V(s)$

```python
for freq_count, s in sim.iter_freqs(freqs, log_skip=50):
    jk_0 = s/c    
     
    Z = sim.impedance(s)
    V = sim.source_plane_wave(e_inc, k_hat*jk_0)
    
    Z_single = Z[0, 0]
    V_single = V[0]
    extinction_single[freq_count] = vdot(V_single, Z_single.solve(V_single)).real
    
    Z_pair, V_pair = Z.combine_parts(V)
    extinction_pair[freq_count] = vdot(V_pair, Z_pair.solve(V_pair)).real
```

Now we plot the extinction cross-section as a function of frequency. We use the cross section to a disc with the same radius as the SRR as a reference to express it in the normalised form of extinction efficiency $Q_{ext}$.

```python
area = pi*(srr.max_distance*0.5)**2
incident = vdot(e_inc, e_inc)/eta_0

plot(freqs*1e-9, extinction_single/incident/area, label='single')
plot(freqs*1e-9, extinction_pair/incident/area, label='pair')
legend(loc='upper right')
axis('tight')
xlabel('freq (GHz)')
ylabel('$Q_{ext}$')
title('Extinction efficiency');
```

This figure shows the fundamental resonance of a single ring. By bringing the two rings near to each other, this resonance splits into two coupled modes. Subsequent examples will show how such modes and be explicitly modelled, and serve as the basis for convenient semi-analytical models of these structures.
